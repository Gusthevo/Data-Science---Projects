```python
base_parlamentares[['uri', 'nome', 'nomeCivil', 'urlRedeSocial', 'urlWebsite', 'municipioNascimento']] = base_parlamentares[['uri', 'nome', 'nomeCivil', 'urlRedeSocial', 'urlWebsite', 'municipioNascimento']].astype('string')
```

```python
base_parlamentares[['dataNascimento', 'dataFalecimento']] = base_parlamentares[['dataNascimento', 'dataFalecimento']].astype('datetime64[ns]')
```

```python
#Retornar o nome dos parlamentares nascidos antes de 1900
for col in base_parlamentares:
    if col == 'dataNascimento':
        print(base_parlamentares[base_parlamentares[col] < '1900-01-01']['nomeCivil'])
```
